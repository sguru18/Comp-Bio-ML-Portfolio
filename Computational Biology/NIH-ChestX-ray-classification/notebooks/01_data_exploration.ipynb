```python
data = pd.read_csv("../data/Data_Entry_2017.csv")
pd.set_option("display.max.columns", None)
data.head()
```

```python
# Distribution of the 15 classes

indices = {
"Atelectasis" : 0,
"Consolidation" : 1,
"Infiltration" : 2,
"Pneumothorax" : 3,
"Edema" : 4,
"Emphysema" : 5,
"Fibrosis" : 6, 
"Effusion" : 7,
"Pneumonia" : 8,
"Pleural_Thickening" : 9,
"Cardiomegaly" : 10,
"Nodule" : 11,
"Hernia" : 12,
"Mass": 13,
"No Finding" : 14,
}

frequencies = [0] * 15
for i in range(data.shape[0]):

    label = data.iloc[i]["Finding Labels"]
    findings = label.split("|")
    for f in findings:
        frequencies[indices[f]] += 1

frequencies = [str(round(f / data.shape[0], 2) * 100) + "%" for f in frequencies]

keys = list(indices.keys())
for i in range(len(indices)):
    print(f'{keys[i]}: {frequencies[i]}')

# Atelectasis: 10.0%
# Consolidation: 4.0%
# Infiltration: 18.0%
# Pneumothorax: 5.0%
# Edema: 2.0%
# Emphysema: 2.0%
# Fibrosis: 2.0%
# Effusion: 12.0%
# Pneumonia: 1.0%
# Pleural_Thickening: 3.0%
# Cardiomegaly: 2.0%
# Nodule: 6.0%
# Hernia: 0.0%
# Mass: 5.0%
# No Finding: 54.0%
```
